```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=np.array([3,8,1,12])
ypoints=np.array([5,6,7,9])
plt.plot(xpoints,ypoints,marker = 'x',ms=10,color='red')
plt.show()
```

```python
ypoints=np.array([1,2,10,4])
plt.plot(ypoints,marker = '*',ms=20,mec='k',mfc='k')
plt.show()
```

```python
ypoints=np.array([4,7,18,9])
plt.plot(ypoints,linestyle='dashed',mec='g',mfc='g',linewidth=5,marker='*',ms=15)
plt.show()
```

```python
ypoints=np.array([6,8,3,10])
plt.plot(ypoints,marker='+',ms=20,mec='k',mfc='k',linestyle='dotted',linewidth=1)
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y,linewidth=2,color='g')
plt.xlabel('average pulse',color='k',fontsize=30)
plt.ylabel('calorie burnage',color='k',fontsize=25)
plt.title('sample',fontsize=35,color='hotpink')
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.title('sports data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
plt.plot(x,y,color='k')
plt.grid(color='g',linestyle='dotted')
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.title('sports data')
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
plt.plot(x,y)
plt.grid(axis ='x')
plt.show()
```

```python
y=np.array([3,15,50,67])
mylabels=('lal','itp','fee','phy')
mycolors=('b','r','hotpink','green')
plt.title('IIITA',color='k',fontsize=25)
plt.pie(y,labels=mylabels,colors=mycolors,startangle=45)
plt.show()
```
